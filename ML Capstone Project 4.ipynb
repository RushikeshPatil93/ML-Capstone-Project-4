# lets use decision tree model it on mic_df

```python
from sklearn import preprocessing

IV = preprocessing.StandardScaler().fit(IV).transform(IV)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(IV,DV, test_size = 0.3, random_state = 15)
```

```python
print(f"Training data shape :: {x_train.shape} | {y_train.shape}")
print(f"Testing data shape :: {x_test.shape} | {y_test.shape}")
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
DTR = DecisionTreeRegressor(criterion='squared_error', splitter='best', max_depth=3)
```

```python
DTR.fit(x_train, y_train)
```

```python
y_Hat = DTR.predict(x_test)
```

```python
f"{DTR.score(x_train, y_train)*100:.2f}%" 
```

```python
from sklearn.metrics import r2_score
```

```python
r2 = r2_score(y_test, y_Hat)
r2
```

```python
best_r2_score = -float('inf')
best_random_state = 0
best_max_depth = 0

for random_state in range(1, 151):
    x_train, x_test, y_train, y_test = train_test_split(IV, DV, test_size=0.2, random_state=random_state)
    
    for max_depth in range(1, 21):
        DTR = DecisionTreeRegressor(criterion='squared_error', splitter='best', max_depth=max_depth)
        DTR.fit(x_train, y_train)
        
        y_pred = DTR.predict(x_test)
        
        current_r2_score = r2_score(y_test, y_pred)
        
        if current_r2_score > best_r2_score:
            best_r2_score = current_r2_score
            best_random_state = random_state
            best_max_depth = max_depth

print("Best R-squared score:", best_r2_score)
print("Random State with the best score:", best_random_state)
print("Max Depth with the best score:", best_max_depth)
```

```python
x_train, x_test, y_train, y_test = train_test_split(IV, DV, test_size=0.2, random_state=31)

DTR = DecisionTreeRegressor(criterion='squared_error', splitter='best', max_depth=3)
DTR.fit(x_train, y_train)
y_pred = DTR.predict(x_test)

micdf_best_acc = round(r2_score(y_test, y_pred) * 100, 2)
```
